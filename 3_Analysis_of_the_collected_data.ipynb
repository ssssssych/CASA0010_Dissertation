# Analysis of the Collected Data

This notebook outlines the steps taken to analyse the collected data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/df.csv")
```

#### Check if it follows a normal distribution, if the variable needs logarithmic transformation 

```python
# Function to plot and calculate skewness and kurtosis
def plot_and_check(df, column):
    f, ax = plt.subplots(1, 2, figsize=(12, 4))
    
    # Before transformation
    sns.histplot(df[column], kde=True, ax=ax[0])
    ax[0].set_title(f'Original Distribution of {column}')
    
    # Apply log transformation
    log_transformed = np.log1p(df[column])
    sns.histplot(log_transformed, kde=True, ax=ax[1])
    ax[1].set_title(f'Log-transformed Distribution of {column}')
    
    plt.tight_layout()
    plt.show()
    
    # Print skewness and kurtosis
    original_skew = df[column].skew()
    original_kurt = df[column].kurt()
    transformed_skew = log_transformed.skew()
    transformed_kurt = log_transformed.kurt()
    
    print(f"{column}:\nOriginal Skewness: {original_skew}, Log-transformed Skewness: {transformed_skew}")
    print(f"Original Kurtosis: {original_kurt}, Log-transformed Kurtosis: {transformed_kurt}\n")

# Check columns
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
for col in numerical_columns:
    plot_and_check(df, col)
```
