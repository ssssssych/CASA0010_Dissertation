# Analysis of the Collected Data

This notebook outlines the steps taken to analyse the collected data.

```python
import pandas as pd
```

```python
df = pd.read_csv("data/df.csv")
```

```python
# List of columns to treat as numerical
numerical_columns = [
    'BldgArea', 'NumBldgs', 'NumFloors','ProxCode',
    'Largest Property Use Type_Gross Floor Area ',
    'Building Age','ENERGY STAR SCORE', 'Weather Normalized Source EUI (kBtu/ft²)',
    'Population with bachelor’s degree or higher', 'Population with 65 years and over', 
    'School Employees', 'Total households', 'Population', 'Mean household income (dollars)', 
    'Civilian noninstitutionalized population with private insurance', 'Median Gross Rent (dollars)',
    'Vacant housing units'
]
# List of columns to treat as categorical
categorical_columns = ['LotType', 'ProxCode', 'BoroCode', 'LandUse']

# Summary statistics for selected numerical variables
num_summary = df[numerical_columns].describe()

# Count and Percentage for each selected categorical variable
cat_count = df[categorical_columns].apply(lambda x: x.value_counts())
cat_percentage = df[categorical_columns].apply(lambda x: x.value_counts(normalize=True) * 100)

# Print or export the results as desired
print("Numerical Summary:")
print(num_summary)

print("\nCount of values for each selected categorical variable:")
print(cat_count)

print("\nPercentage of values for each selected categorical variable:")
print(cat_percentage)
```
